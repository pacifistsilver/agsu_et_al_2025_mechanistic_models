# mRNA Count Probability Distributions for Nanog and Sox2
This notebook loads the 129 and CAST allele datasets, extracts the counts for Nanog and Sox2, and plots their probability distributions (histograms).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Known ENSMUST IDs
nanog_ids = ['ENSMUST00000112581', 'ENSMUST00000112580', 'ENSMUST00000012540', 'ENSMUST00000553288', 
             'ENSMUST00000553289', 'ENSMUST00000553290', 'ENSMUST00000553291', 'ENSMUST00000553292', 
             'ENSMUST00000553293', 'ENSMUST00000553294', 'ENSMUST00000553295', 'ENSMUST00000553296', 
             'ENSMUST00000553297', 'ENSMUST00000578199']
sox2_ids = ['ENSMUST00000099151']

def extract_gene_counts(filepath, gene_name, transcript_ids):
    print(f"Processing {gene_name} in {filepath}...")
    df = pd.read_csv(filepath, sep=' ', index_col=0)
    available_ids = [tid for tid in transcript_ids if tid in df.index]
    if not available_ids:
        print(f"No transcripts found for {gene_name} in {filepath}")
        return np.array([])
    
    # Sum counts across all available transcripts for this gene
    counts = df.loc[available_ids].sum(axis=0)
    return counts.values
```

```python
nanog_129 = extract_gene_counts('data/ochiai/GSE132589_ASEcount_G1_129.txt', 'Nanog', nanog_ids)
nanog_cast = extract_gene_counts('data/ochiai/GSE132589_ASEcount_G1_CAST.txt', 'Nanog', nanog_ids)

sox2_129 = extract_gene_counts('data/ochiai/GSE132589_ASEcount_G1_129.txt', 'Sox2', sox2_ids)
sox2_cast = extract_gene_counts('data/ochiai/GSE132589_ASEcount_G1_CAST.txt', 'Sox2', sox2_ids)

print(f"Nanog counts shape - 129: {nanog_129.shape}, CAST: {nanog_cast.shape}")
print(f"Sox2 counts shape - 129: {sox2_129.shape}, CAST: {sox2_cast.shape}")
```

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

def plot_hist(ax, data, title, color):
    if len(data) == 0:
        ax.set_title(f"{title} (No Data)")
        return
    ax.hist(data, bins=30, density=True, color=color, alpha=0.7, edgecolor='black')
    ax.set_title(title, fontsize=14)
    ax.set_xlabel('mRNA Counts', fontsize=12)
    ax.set_ylabel('Probability', fontsize=12)

plot_hist(axes[0, 0], nanog_129, 'Nanog (129 Allele)', 'dodgerblue')
plot_hist(axes[0, 1], nanog_cast, 'Nanog (CAST Allele)', 'orange')
plot_hist(axes[1, 0], sox2_129, 'Sox2 (129 Allele)', 'dodgerblue')
plot_hist(axes[1, 1], sox2_cast, 'Sox2 (CAST Allele)', 'orange')

plt.tight_layout()
plt.show()
```
